In this part we will take a look at different ways to customize our data graphs.

There are several ways to customize graphs in the seaborn library. We will take a look at two of the most practical ones:

- Using Seaborn function parameters
- Using matplotlib's pyplot

We will use these together. Let's go!

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Once again, we will work with the automobile data set that we used for the data viz tutorial. And again,  we will change the values of the origin column to actual region names. Also, just like before, we will turn '?' into NaN so that they will not change column data types to strings.

```python
# Load the data frame, turn '?' entries into NaN
auto_df = pd.read_csv("auto-mpg.csv", na_values='?')

# Replace region numbers with actual region names
auto_df.origin = auto_df.origin.replace({1:'USA', 2:'Europe', 3:'Asia'})
auto_df.head()
```

## Titles & Axis Labels

By using matplotlib's pyplot (imported as plt), we can:

* Write a title for our visualization
* Write labels for x & y variables

```python
sns.scatterplot(x=auto_df.displacement, y= auto_df.mpg)
plt.title("Displacement-MPG Relation")
plt.xlabel("Engine displacement")
plt.ylabel("Miles-per-gallon ratio")
plt.show()
```

## Annotate Text with Pointers

By using pyplot, we can put annotations on our visualization.

In the following example, we show a point in our viz with a red pointer and we add a note to attract attention to and explain a certain point. The parameters are explained as follows:

* text -- The text we want to write
* xy -- the point we want to annotate
* xytext -- the point where we want to put the text
* arrowprops -- takes a dictionary of values for the pointer. Here we used green as its color.

```python
sns.scatterplot(x=auto_df.displacement, y= auto_df.mpg)
# text, x-y meeting point for pointer, location of the text, color of the pointer
plt.annotate(text="Important", xy=(200, 30), xytext=(210, 40), arrowprops={'color':'green'})
plt.show()
```

## Draw Lines

We can use pyplot to draw lines from axes.

* axhline is for horizontal lines
* axvline is for vertical lines

In the following examples, we draw lines from both the x and y axes, higlighting the means of two variables.

```python
sns.scatterplot(x=auto_df.displacement, y= auto_df.mpg)
plt.axhline(np.mean(auto_df.mpg), color = 'red')
plt.axvline(np.mean(auto_df.displacement), color = 'green')
plt.show()
```

## Set Hue & Size

We can use Seaborn's hue and size parameters to quickly make our graphs more informative. These parameters also add a legend to the graph that is quite self-explanatory. 

```python
sns.scatterplot(x=auto_df.displacement, y= auto_df.mpg, hue=auto_df.origin, size=auto_df.cylinders)
plt.show()
```

In the graph above, Seaborn automatically painted data points based on their origin and adjusted their size in accordance with the number of cylinders. 

## Change Orientation (Vertical or Horizontal)

We can use Seaborn functions' orient= parameter to change the orientation of our graphs.

```python
# Vertical (default)
sns.barplot(x=auto_df.origin, y=auto_df.mpg, ci = None) # Do not show confidence interval
plt.show()
```

```python
# Horizontal (x must be a numeric variable)
sns.barplot(x=auto_df.mpg, y=auto_df.origin, orient='h', ci=None) 
plt.show()
```
